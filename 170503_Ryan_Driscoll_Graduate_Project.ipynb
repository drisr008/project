## Using Python to create 'Instagram-like' image filters

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
coffee = Image.open('coffee.jpg')
#print (coffee.mode)

print ('coffee (RGB)')
plt.imshow(coffee)
plt.show()

r, g, b, = coffee.split()

print ('red channel')
plt.imshow(r)
plt.show()

print ('green channel')
plt.imshow(g)
plt.show()

print ('blue channel')
plt.imshow(b)
plt.show()
```

```python
texture = Image.open('texture.jpg')
#print (texture.mode)

print ('texture (RGB)')
plt.imshow(texture)
plt.show()

rT, gT, bT, = texture.split()

print ('red channel')
plt.imshow(rT)
plt.show()

print ('green channel')
plt.imshow(gT)
plt.show()

print ('blue channel')
plt.imshow(bT)
plt.show()
```

```python
new_image = Image.merge('RGB', (r, g, bT))
plt.imshow(new_image)
plt.xticks([]), plt.yticks([])
plt.show()
```
